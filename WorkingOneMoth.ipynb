```python
quartiles = Data2021OneDay['tripdurationMin'].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

output=f'''
    -------------------------------------------------------
    The lower quartile of trip duration is: {round(lowerq,2)}
    
    The upper quartile of trip duration is: {round(upperq,2)}
    
    The interquartile range of trip duration is: {round(iqr,2)}
    
    The the median of trip duration is: {round(quartiles[0.5],2)}
    
    Values below {round(lower_bound,2)} could be outliers.
    
    Values above {round(upper_bound,2)} could be outliers.

    '''
print(output)
```

```python
Data2021Clean=Data2021OneDay.loc[Data2021OneDay['tripdurationMin']<upper_bound]
Data2021Outliers=Data2021OneDay.loc[Data2021OneDay['tripdurationMin']>upper_bound]
Data2021Clean['tripdurationMin'].plot(kind='box')
```

```python
  
#generate DataFrame for Stations
Stations=pd.DataFrame({
        'station id':[i for i in Data2021Clean['start station id']],
        'station name':[i for i in Data2021Clean['start station name']],
        'station latitude':[i for i in Data2021Clean['start station latitude']],
        'station longitude':[i for i in Data2021Clean['start station longitude']]
})

Stations.drop_duplicates(inplace=True)
Stations
    
```

```python
Category=[]
for item in Data2021Clean['Age']:
    if item<12:
        string='Child'
    elif item<19:
        string='Teen'
    elif item<35:
        string='Young Adult'
    elif item<55:
        string='Adult'
    elif item<105:
        string='Senior'
    else:
        string='Not category'
    Category.append(string)
Data2021Clean['Age category']=Category
Data2021Clean.head()
```

```python
TripCategory=[]
for start, end in zip(Data2021Clean['start station id'],Data2021Clean['end station id']):
    if start==end:
        string='loop'
    else:
        string='simple trip'
    TripCategory.append(string)
Data2021Clean['Trip category']=TripCategory
Data2021Clean.head()
```

```python
def DateWeek(stringInput):
    
    weekDays = ("Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday")
    
    stringDate, stringTime=stringInput.split(' ')
    stringYear, stringMonth, stringDay=stringDate.split('-')
    
    dateResult=date(int(stringYear),int(stringMonth),int(stringDay))
    Result=weekDays[dateResult.weekday()]
    
    return Result
```

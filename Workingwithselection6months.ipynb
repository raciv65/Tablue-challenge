```python
Files=[]
for path, directories, files in walk(pathFiles):
    Files.append(files)
Files
```

```python
Year2020=[]

for i in range(len(Files[0])):
    string=Files[0][i]
    if string.find('2020')!=-1:
        if string.find('JC')==-1:
            Year2020.append(string)

Year2020
```

```python
def DataYear(List):
    if isinstance(List,list):
        Data=[]
        Errors=[]
        for element in range(len(List)):
            try:
                Data.append(pd.read_csv(os.path.join(pathFiles,List[element]), sep=','))
            except:
                Errors.append(f'cheouk out file: {List[element]}')
        
        return Data, Errors
    else:
        return 'Not a list'
```

```python
Data2020, Error2020=DataYear(Year2020)
```

```python
quartiles = CombinedYearsdfDay['tripdurationMin'].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

output=f'''
    -------------------------------------------------------
    The lower quartile of trip duration is: {round(lowerq,2)}
    
    The upper quartile of trip duration is: {round(upperq,2)}
    
    The interquartile range of trip duration is: {round(iqr,2)}
    
    The the median of trip duration is: {round(quartiles[0.5],2)}
    
    Values below {round(lower_bound,2)} could be outliers.
    
    Values above {round(upper_bound,2)} could be outliers.

    '''
print(output)
```

```python
CombinedYearsClean=CombinedYearsdfDay.loc[CombinedYearsdfDay['tripdurationMin']<upper_bound]
CombinedYearsOutsider=CombinedYearsdfDay.loc[CombinedYearsdfDay['tripdurationMin']>upper_bound]
CombinedYearsClean['tripdurationMin'].plot(kind='box')
```
